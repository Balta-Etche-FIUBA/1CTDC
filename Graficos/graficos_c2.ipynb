```python
def graficar_variable(data, x_col, y_col, x_label=None, y_label=None, title=None):
    """
    Genera un gráfico de línea con los valores máximos y mínimos resaltados.
    
    Parámetros:
    - data: DataFrame que contiene los datos.
    - x_col: Nombre de la columna a usar en el eje X.
    - y_col: Nombre de la columna a usar en el eje Y.
    - x_label: Etiqueta del eje X (opcional, por defecto usa el nombre de la columna).
    - y_label: Etiqueta del eje Y (opcional, por defecto usa el nombre de la columna).
    - title: Título del gráfico (opcional).
    """
    max_index = data[y_col].idxmax()
    min_index = data[y_col].idxmin()
    
    max_x = data[x_col][max_index]
    max_y = data[y_col][max_index]
    min_x = data[x_col][min_index]
    min_y = data[y_col][min_index]
    
    plt.figure(figsize=(8, 5))
    plt.plot(data[x_col], data[y_col], label=y_col, color='b', linewidth=2)
    
    plt.scatter(max_x, max_y, color='r', label="Máximo", zorder=3)
    plt.annotate(f"Máx: ({max_x:.2f}, {max_y:.4f})", 
                 (max_x, max_y), 
                 textcoords="offset points", 
                 xytext=(10, -10), 
                 ha='left', 
                 fontsize=12, 
                 color='r')
    
    plt.scatter(min_x, min_y, color='g', label="Mínimo", zorder=3)
    plt.annotate(f"Mín: ({min_x:.2f}, {min_y:.4f})", 
                 (min_x, min_y), 
                 textcoords="offset points", 
                 xytext=(-30, 10), 
                 ha='right', 
                 fontsize=12, 
                 color='g')
    
    plt.xlabel(x_label if x_label else x_col, fontsize=14)
    plt.ylabel(y_label if y_label else y_col, fontsize=14)
    plt.title(title if title else f"Relación entre {x_col} y {y_col}", fontsize=16)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(fontsize=12)
    plt.show()
```
